```python
import re
import datetime
import requests
import pandas as pd
from bs4 import BeautifulSoup
from datetime import timedelta,date
```

```python
results = requests.get('https://www.graphic.com.gh/tech-news.html')
results.url
```

```python
results
```

```python
soup = BeautifulSoup(results.content, 'lxml')
soup
```

```python
containers = soup.findAll('div', {'class': 'separated-item-2'})
containers
```

```python
d_title=[]
d_article_url=[]
d_author=[]
d_postdate=[]


base_url = "https://www.graphic.com.gh"
for container in containers:
    title = container.article.h2.a.text
    article_url = base_url + (container.article.h2.a["href"])
    get_author = container.div.find('div', {'class': 'bd-posticonauthor-16'})
    author = get_author.span.span.text.split('BY: ')[1]
    post_time = container.div.find('div', {'class': 'bd-posticondate-15'})
    postdate = post_time.span.span.time['datetime']
    
    d_title.append(title)
    d_article_url.append(article_url)
    d_author.append(author)
    d_postdate.append(postdate)


    df = pd.DataFrame({
        'Title': d_title,
        'URL': d_article_url,
        'Author' : d_author,
        'PostDate': d_postdate

    })
df
```
