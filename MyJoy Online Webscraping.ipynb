```python
import re
import datetime
import requests
import pandas as pd
from bs4 import BeautifulSoup
from datetime import timedelta,date
```

```python
results = requests.get('https://www.myjoyonline.com/category/business/technology-business/')
results.url
```

```python
results
```

```python
soup = BeautifulSoup(results.content, 'lxml')
soup
```

```python
d_title=[]
d_article_url=[]
d_category=[]
d_author=[]
d_postdatetime=[]


containers = soup.findAll('div', {'class': 'col-lg-9'})
for container in containers:
    for row in container.findAll('div', {'class': 'row'}):
        for individual_post in row.findAll('div', {'class': 'col-xs-6'}):
            post_url = individual_post.div.div.a['href']
            post_results = requests.get(post_url)
            post_soup = BeautifulSoup(post_results.content, 'lxml')
            single_post_container = post_soup.findAll('div', {'class': 'container'})
            for single_post in single_post_container:
                for post in single_post.findAll('div', {'class': 'col-xl-12'}):
                    title = post.find('div', {'class': 'article-title'}).a.h1.text.strip('\n')
                    find_author = post.find('div', {'class': 'article-meta'}).a
                    if(find_author):
                        author = find_author.text
                    else:
                        author = "No Author"
                    postdatetime = post.find('div', {'class': 'article-meta'}).div.text
                    category = post.div.text
#                     print(post_url, title, author, category, postdatetime)
                    
                    d_title.append(title)
                    d_article_url.append(post_url)
                    d_category.append(category)
                    d_author.append(author)
                    d_postdatetime.append(postdatetime)


                    df = pd.DataFrame({
                        'Title': d_title,
                        'URL': d_article_url,
                        'Category' : d_category,
                        'Author': d_author,
                        'PostDate': d_postdatetime

                    })
df          
```
